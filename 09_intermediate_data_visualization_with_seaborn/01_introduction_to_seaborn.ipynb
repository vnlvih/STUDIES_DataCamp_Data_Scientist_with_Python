```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
red = pd.read_csv("./datasets/winequality-red.csv",sep=';')
red['type'] = 'red'
white = pd.read_csv("./datasets/winequality-white.csv",sep=';')
white['type'] = 'white'
red.shape,white.shape
```

```python
df = pd.concat([red,white])
```

### Matplotlib
- provides the raw building blocks for Seaborn's visualizations
- It can also be used on its own to plot data

```python
fig, ax = plt.subplots()
ax.hist(df['alcohol'])
plt.show()
```

### Pandas

- Its also supports basic plottting capability

```python
df['alcohol'].plot.hist()
```

### Seaborn
- supports complex visualizations of data
-It's built on matplotlib and works best with panda's dataframes

```python
sns.histplot(df['alcohol'])
plt.show()
```

***displot***

- levarages the `histplot` and other functions for distribution plots
- by default, it generates a histogram but can also generate other plot types

```python
sns.displot(df['alcohol'])
plt.show()
```

```python
sns.displot(df['alcohol'], kind = 'kde')
plt.show()
```

#### Creating a histogram

```python
#you can overlay a KDE plot on the histogram and specify the number of bins to use

sns.displot(df['alcohol'], kde=True, bins=10)
plt.show()
```

```python
# a rug plot is an alternative way to view the distribution of data by including small tickmarks along the x axis

sns.displot(df['alcohol'], kind='kde', rug=True,fill=True)
plt.show()
```

***Further plot types***
- The `displot` function uses several functions including `kdeplor`, `rugplot` and `ecdfplot`

```python
# ecdfplot shows the cumulative distribution of the data

sns.displot(df['alcohol'], kind='ecdf')
plt.show()
```

### Regression plots

```python
sns.regplot(data=df, x="alcohol",y="pH")
plt.show()
```

```python
# low level
sns.regplot(data=df, x="alcohol",y="quality")
plt.show()
```

```python
# high level
sns.lmplot(data=df, x="alcohol",y="quality")
plt.show()
```

```python
#organize data by colors

sns.lmplot(data=df,
           x="quality",
           y="alcohol",
           hue='type')
plt.show()
```

```python
#organize data by columns

sns.lmplot(data=df,
           x="quality",
           y="alcohol",
           col="type")
plt.show()
```
