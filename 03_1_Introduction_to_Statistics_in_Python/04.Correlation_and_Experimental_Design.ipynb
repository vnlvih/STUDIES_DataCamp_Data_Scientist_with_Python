## Correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
msleep = pd.read_csv("./datasets/msleep.csv")
```

```python
sns.scatterplot(x = "sleep_total", y= "sleep_rem", data=msleep)
plt.show()
```

```python
sns.lmplot(data=msleep, x= "sleep_total", y="sleep_rem", ci=None)
plt.show()
```

```python
#computing correlation

msleep["sleep_total"].corr(msleep['sleep_rem'])
```

```python
msleep["sleep_rem"].corr(msleep['sleep_total'])
```

### Non-linear relationships

```python
msleep['bodywt'].corr(msleep['awake'])
```

```python
sns.scatterplot(x = "bodywt", y= "awake", data=msleep)
plt.show()
```

```python
#Distribution of body weight


plt.hist(msleep['bodywt'])
plt.grid(True)
plt.show()
```

When data is highly skewed like this, we can apply a log transformation.

```python
#log transformation
import numpy as np
msleep['log_bodywt'] = np.log(msleep['bodywt'])

sns.lmplot(x="log_bodywt",
           y = 'awake',
           data=msleep,
           ci=None)
plt.show()
print(msleep['log_bodywt'].corr(msleep['awake']))
```
